## Grid supply, Voltage (V) have many outliers and Overload Condition, Transformer Fault have one outlier

### Handling outliers

```python
minimum,Q1,median,Q3,maximum=np.quantile(df['Grid Supply (kW)'],[0,0.25,0.50,0.75,1.0])
```

```python
IQR=Q3-Q1
print(minimum,Q1,median,Q3,maximum,IQR)
```

```python
lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
```

```python
avg=np.average(df['Grid Supply (kW)'])
```

```python
outlier=[]
for i in range(len(df['Grid Supply (kW)'])):
    if df['Grid Supply (kW)'].iloc[i]<lower_fence or df['Grid Supply (kW)'].iloc[i]> upper_fence :
        outlier.append(df['Grid Supply (kW)'].iloc[i])
print(len(outlier),outlier)
```

```python
for i in range(len(df['Grid Supply (kW)'])):
    if df['Grid Supply (kW)'].iloc[i]<lower_fence or df['Grid Supply (kW)'].iloc[i]> upper_fence :
        df['Grid Supply (kW)'].iloc[i]=avg
```

```python
plt.boxplot(df['Grid Supply (kW)'])
```

# Grid outliers handled

```python
minimum,Q1,median,Q3,maximum=np.quantile(df['Voltage (V)'],[0,0.25,0.50,0.75,1.0])
IQR=Q3-Q1
print(minimum,Q1,median,Q3,maximum,IQR)
lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
avg=np.average(df['Voltage (V)'])
for i in range(len(df['Voltage (V)'])):
    if df['Voltage (V)'].iloc[i]<lower_fence or df['Voltage (V)'].iloc[i]> upper_fence :
        df['Voltage (V)'].iloc[i]=avg
```

# Voltage (V) outliers handled

```python
y=df['Predicted Load (kW)']
X=df[df.columns[df.columns!=y.name]]
```

```python
y.head()
```

```python
X.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.05,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
xtrain_scaled=scaler.fit_transform(x_train)
xtest_scaled=scaler.transform(x_test)
```

```python
import pickle
pickle.dump(scaler,open("Load_forecasting_scaler.pkl","wb"))
```

```python
from sklearn.linear_model import ElasticNet,LinearRegression
from sklearn.metrics import r2_score
```

```python
elastic=ElasticNet()
regressor=LinearRegression()
```

```python
elastic.fit(xtrain_scaled,y_train)
regressor.fit(xtrain_scaled,y_train)
y_pred_elastic=elastic.predict(xtest_scaled)
y_pred_regressor=regressor.predict(xtest_scaled)
print("R2 Score Elastic=",abs(r2_score(y_pred_elastic,y_test))*100)
print("R2 Score Regressor=",abs(r2_score(y_pred_regressor,y_test))*100)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
```

```python
Random_Forest=RandomForestRegressor()
Decision_Tree=DecisionTreeRegressor()
```

```python
Random_Forest.fit(xtrain_scaled,y_train)
Decision_Tree.fit(xtrain_scaled,y_train)
```

```python
y_pred_RF=Random_Forest.predict(xtest_scaled)
y_pred_DT=Decision_Tree.predict(xtest_scaled)
```

```python
print("R2 Score RF=",abs(r2_score(y_test,y_pred_RF))*100)
print("R2 Score DT=",abs(r2_score(y_test,y_pred_DT))*100)
```

```python
pickle.dump(regressor,open("rergessor.pkl","wb"))
pickle.dump(elastic,open("elasticnet.pkl","wb"))
pickle.dump(Decision_Tree,open("descision_tree.pkl","wb"))
pickle.dump(Random_Forest,open("random_forest.pkl","wb"))
```
